```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(24,16)})
%matplotlib inline
```

```python
df=pd.read_csv("salary.csv")
numeric=pd.read_csv("salary_numeric.csv")
numeric
```

```python
sns.displot(data=df, x='salary', kind="kde",  hue='industry')
```

```python
sns.scatterplot(data=df, x='industry_experience', y='salary', hue='gender')
```

```python
df_male = df.drop(df[df.gender == 'Female'].index)
df_female = df.drop(df[df.gender == 'Male'].index)
```

```python
df_male = df_male.drop('gender', axis=1)
df_female = df_female.drop('gender', axis=1)
df_male = df_male.drop('industry', axis=1)
df_female = df_female.drop('industry', axis=1)
```

```python
sns.heatmap(data=df_male.corr(), cmap='RdBu_r', annot=True)
```

```python
sns.heatmap(data=df_female.corr(), cmap='RdBu_r', annot=True)
```
