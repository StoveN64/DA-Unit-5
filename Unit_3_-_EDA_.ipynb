```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = Higher education level leads to higher salary

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df1.salary, df1.education)
# pearsonr(df['salary'], df['age'])
```

[interpret result]
> Middling correlation

#### Are these results statistically significant? How do you know?

Not a huge impact, but I hypothesize that its due to the diminishing returns seen in the charts above.

### What is the correlation between salary and age?

#### Your Hypothesis: Higher age value leads to higher pay

```python
# calculate correlation and pvalue of salary and age with pearsonr method
spearmanr(df['salary'], df['industry_experience'])
```

[interpret result]
> Larger correlation

#### Are these results statistically significant? How do you know? 

Now we are gettting somewhere. Since theres not as many variable difference as above, we see more significance between higher age and higher pay.

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = Should see higher salary correlate with more industry experience

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
spearmanr(df['salary'], df['industry_experience'])
```

[interpret result]
> High correlation, not as much as age

#### Are these results statistically significant? How do you know?

They are statistically significant, BUT salary is different in each industry, so we wont see a higher correlation as we did with age, which is more typically common for higher salaries.
